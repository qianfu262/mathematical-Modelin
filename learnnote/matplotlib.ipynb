# matplotlib的使用

```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore
import pandas as pd # type: ignore
# 用拟合绘制一个正弦函数
x=np.linspace(0,2*np.pi,100) #表示0到10之间的100个随机数
print(x)
y=np.sin(x)
plt.plot(x,y)
plt.title('y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
## 加入各种元素
x2=np.linspace(0,2*np.pi,1000)
y2=np.sin(x2)
plt.scatter(x,y,marker='*',color='r',label='数据集')#用散点表示数据,marker表示形状，color表示颜色
plt.plot(x2,y2,linestyle='--',label='拟合曲线')
plt.show()
```

```python
# 绘制两个图表
fig,test=plt.subplots(1,2) #表示1行2列的图表
test[0].scatter(x,y,marker='*',color='r',label='数据集')
test[0].set_xlabel('x1')
test[0].set_ylabel('y')
test[1].plot(x2,y2,linestyle='--',label='拟合曲线')
test[1].set_xlabel('x2')
plt.show()
```

```python
# 绘制柱状图
x=[1,2,3]
y=[2,4,10]
plt.bar(x,y)
plt.show()
```
